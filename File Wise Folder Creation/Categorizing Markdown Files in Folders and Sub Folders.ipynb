````python
# Cell 4: Create a function to handle files not in our mapping
def auto_categorize(filename, content=None):
    """
    Attempt to auto-categorize files not in our mapping.
    This is a fallback function.
    """
    filename_lower = filename.lower()
    
    # Check for patterns
    if "dax-" in filename_lower:
        return ("Concepts", "Miscellaneous")
    elif "-functions-dax" in filename_lower:
        return ("Categories-Overview", None)
    elif filename_lower.startswith("is"):
        return ("Core Functions", "Information")
    elif any(keyword in filename_lower for keyword in ["date", "time", "day", "month", "year"]):
        return ("Core Functions", "Date-Time")
    elif any(keyword in filename_lower for keyword in ["sum", "count", "average", "max", "min"]):
        return ("Core Functions", "Aggregation")
    elif any(keyword in filename_lower for keyword in ["filter", "all", "selected"]):
        return ("Core Functions", "Filter")
    elif any(keyword in filename_lower for keyword in ["sin", "cos", "tan", "log", "exp"]):
        return ("Core Functions", "Math-Trig")
    
    # Default fallback
    return ("Core Functions", "Other")
````

````python
# Cell 6: Analyze and categorize all files
def analyze_files(base_path, categorization):
    """
    Analyze all markdown files and determine their categorization.
    Returns a DataFrame with file details.
    """
    # Get all markdown files
    md_files = list(base_path.glob("*.md"))
    print(f"Found {len(md_files)} markdown files")
    
    file_data = []
    uncategorized_files = []
    
    for file_path in md_files:
        filename = file_path.name
        
        # Check if file is in our categorization mapping
        if filename in categorization:
            category, subcategory = categorization[filename]
            source = "Mapping"
        else:
            # Try to auto-categorize
            category, subcategory = auto_categorize(filename)
            source = "Auto"
            uncategorized_files.append(filename)
        
        file_data.append({
            "filename": filename,
            "current_path": str(file_path),
            "category": category,
            "subcategory": subcategory,
            "source": source
        })
    
    # Create DataFrame
    df = pd.DataFrame(file_data)
    
    print(f"\nCategorization Summary:")
    print(f"  Total files: {len(df)}")
    print(f"  Mapped files: {len(df[df['source'] == 'Mapping'])}")
    print(f"  Auto-categorized files: {len(df[df['source'] == 'Auto'])}")
    print(f"  Uncategorized files (needed auto): {len(uncategorized_files)}")
    
    if uncategorized_files:
        print(f"\nFiles that needed auto-categorization:")
        for file in uncategorized_files[:20]:
            print(f"  - {file}")
        if len(uncategorized_files) > 20:
            print(f"  ... and {len(uncategorized_files) - 20} more")
    
    return df, uncategorized_files

# Analyze files
df_files, uncategorized = analyze_files(base_path, categorization)
````

````python
# Cell 9: Function to actually move the files
def organize_files(df_files, base_path, backup=True):
    """
    Organize files into the categorized folder structure.
    """
    moved_files = []
    skipped_files = []
    error_files = []
    
    # Create backup folder if requested
    if backup:
        backup_path = base_path / "backup_original"
        backup_path.mkdir(exist_ok=True)
        print(f"Created backup folder: {backup_path}")
    
    total_files = len(df_files)
    print(f"\nStarting to organize {total_files} files...")
    print("=" * 80)
    
    for i, row in df_files.iterrows():
        current_path = Path(row['current_path'])
        filename = row['filename']
        category = row['category']
        subcategory = row['subcategory']
        
        # Determine new path
        if pd.isna(subcategory):
            new_path = base_path / category / filename
        else:
            new_path = base_path / category / subcategory / filename
        
        try:
            # Skip if file doesn't exist (shouldn't happen)
            if not current_path.exists():
                skipped_files.append((filename, "Source file not found"))
                continue
            
            # Skip if source and destination are the same
            if current_path == new_path:
                skipped_files.append((filename, "Already in correct location"))
                continue
            
            # Create backup if requested
            if backup:
                backup_file_path = backup_path / filename
                shutil.copy2(current_path, backup_file_path)
            
            # Create destination directory if it doesn't exist
            new_path.parent.mkdir(parents=True, exist_ok=True)
            
            # Move the file
            shutil.move(str(current_path), str(new_path))
            
            moved_files.append({
                'filename': filename,
                'from': str(current_path),
                'to': str(new_path),
                'category': category,
                'subcategory': subcategory if not pd.isna(subcategory) else 'None'
            })
            
            # Progress indicator
            if (i + 1) % 50 == 0:
                print(f"Processed {i + 1}/{total_files} files...")
                
        except Exception as e:
            error_files.append((filename, str(e)))
            print(f"Error moving {filename}: {e}")
    
    print("\n" + "=" * 80)
    print("ORGANIZATION COMPLETE")
    print("=" * 80)
    print(f"Total files processed: {total_files}")
    print(f"Successfully moved: {len(moved_files)}")
    print(f"Skipped: {len(skipped_files)}")
    print(f"Errors: {len(error_files)}")
    
    if backup:
        print(f"\nBackup created in: {backup_path}")
    
    if error_files:
        print("\nFiles with errors:")
        for filename, error in error_files[:10]:
            print(f"  - {filename}: {error}")
        if len(error_files) > 10:
            print(f"  ... and {len(error_files) - 10} more")
    
    return moved_files, skipped_files, error_files

# UNCOMMENT THE NEXT LINE TO ACTUALLY MOVE FILES
# moved, skipped, errors = organize_files(df_files, base_path, backup=True)
````
