```python
import os
import mammoth
```

```python
OUTPUT_ROOT = r"C:\Users\BalasubramanianPG\Videos\Captures"
```

```python
def clean_name(filename):
    name, _ = os.path.splitext(filename)
    name = name.replace("_", " ")
    name = " ".join(word.capitalize() for word in name.split())
    return name + ".md"
```

```python
def convert_docx_to_md(input_path, output_path):
    try:
        with open(input_path, "rb") as docx_file:
            result = mammoth.convert_to_markdown(docx_file)
            markdown = result.value

        with open(output_path, "w", encoding="utf-8") as md_file:
            md_file.write(markdown)

        print(f"Converted → {output_path}")

    except Exception as e:
        print(f"Failed: {input_path}")
        print(e)
```

```python
def process_folder(source_root):
    for dirpath, dirnames, filenames in os.walk(source_root):
        for file in filenames:
            if file.lower().endswith(".docx"):
                source_file = os.path.join(dirpath, file)

                relative_path = os.path.relpath(dirpath, source_root)
                output_folder = os.path.join(OUTPUT_ROOT, relative_path)
                os.makedirs(output_folder, exist_ok=True)

                cleaned_filename = clean_name(file)
                output_file = os.path.join(output_folder, cleaned_filename)

                print("--------------------------------------------------")
                print(f"Source: {source_file}")
                print(f"Output: {output_file}")

                convert_docx_to_md(source_file, output_file)
```

```python
SOURCE_ROOT = r"C:\Users\BalasubramanianPG\Downloads\OneDrive_2025-11-27"
process_folder(SOURCE_ROOT)
```
