```python
def create_yaml_header(props):
    """Constructs the YAML block with no extra leading/trailing whitespace."""
    header = "---\n"
    for key, value in props.items():
        # Wrap values in quotes to ensure [[Snowflake]] stays a link 
        # and doesn't turn into [["Snowflake"]]
        if "[[" in str(value) or value == "":
            header += f'{key}: "{value}"\n'
        else:
            header += f"{key}: {value}\n"
    header += "---" 
    return header
```

```python
def update_markdown_files(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.md'):
                file_path = os.path.join(root, file)
                
                # 1. Get the Chapter (the name of the immediate folder)
                chapter_name = os.path.basename(root)
                
                # 2. Get the Date Created (Windows creation time)
                ctime = os.path.getctime(file_path)
                date_created = datetime.datetime.fromtimestamp(ctime).strftime('%Y-%m-%d')
                
                # Define the properties exactly as requested
                properties = {
                    "Chapter": chapter_name,
                    "Link": "",
                    "Cover": "",
                    "Date Created": date_created,
                    "Completed": "",
                    "Area": "",
                    "Sub Area": "[[Snowflake]]" # Linked to your Snowflake page
                }
                
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()

                # Logic to separate existing YAML from the body
                if content.startswith('---'):
                    try:
                        # Split by the YAML delimiters
                        parts = content.split('---')
                        # parts[0] is empty, parts[1] is old YAML, parts[2:] is content
                        body = "---".join(parts[2:]).strip()
                    except (ValueError, IndexError):
                        body = content.strip()
                else:
                    body = content.strip()

                # Combine: New Header + exactly one newline + the cleaned body
                new_content = create_yaml_header(properties) + "\n" + body

                with open(file_path, 'w', encoding='utf-8') as f:
                    f.write(new_content)
                
                print(f"Updated: {file}")

if __name__ == "__main__":
    update_markdown_files(root_dir)
```
